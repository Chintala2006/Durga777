```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

```python
df = pd.read_csv('Mall_Customers.csv')
```

```python
df.head()
```

```python
df_male = df[df.Genre == 'Male']
df_female = df[df.Genre == 'Female']
df_male.head()
```

```python
# Visualization
labels = ['Annual Income (k$)', 'Spending Score (1-100)']
colors = ['blue','red']
sizes = [df_male["Age"].mean(), df_male["Spending Score (1-100)"].mean()] 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  
plt.tight_layout()
plt.title("Men average spending at the mall")
plt.show()
```

```python
# Visualization
labels = ['Annual Income (k$)', 'Spending Score (1-100)']
colors = ['blue','red']
sizes = [df_female["Age"].mean(), df_female["Spending Score (1-100)"].mean()] 

fig1, ax1 = plt.subplots()
ax1.pie(sizes, colors = colors, labels=labels, autopct='%1.1f%%', startangle=90)

centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax1.axis('equal')  
plt.tight_layout()
plt.title("Female average spending at the mall")
plt.show()
```
